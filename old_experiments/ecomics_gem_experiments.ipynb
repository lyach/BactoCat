### ATP Limitation

- Cells must balance ATP supply (production reactions) with ATP demand (biomass formation + maintenance).
- “Limiting ATP” means shrinking the net ATP budget available for growth.

This is achieved by:
1) Raising the ATP drain rxn (ATPM): ATPM represents energy needed for tasks not tied to biomass (ion pumping, turnover, etc.). By setting a larger  flux to this rxn, the model will burn more ATP each hour, leaving less ATP for growth-associated tasks.
2) Capping ATP-generating rxns (ATPS4rpp): Placing a low upper bound limits the maximal ATP that oxidative phosphorylation can supply. The rest of metabolism must now rely on less efficient phosphorylation (at substrate level), increasing metabolic “cost” per biomass unit.

```python
# load
model = load_model("iML1515")

solution = model.optimize()
print('OG solution:')
print(solution.objective_value)
print()

# Raise ATP drain rxn
atp_maint = model.reactions.get_by_id("ATPM")   # non-growth-associated maintenance
atp_maint.lower_bound = 20    # mmol · gDW⁻¹ · h⁻¹ (↑ demand → ATP scarcer)
atp_maint.upper_bound = 20

# Cap the ATP synthase rxn
atp_synth = model.reactions.get_by_id("ATPS4rpp")  # F₁F₀-ATP synthase
atp_synth.upper_bound = 5      # limit ATP supply


solution = model.optimize()
print('NEW solution:')
print(solution.objective_value)
print()
```

### NADH (re oxidation) limitation

- Catabolic (consuming) pathways produce NADH
- Respiratory or fermentative steps (growth) re-oxidize it back to NAD⁺
- Blocking out rxns that turn NADH back into NAD⁺ will limit GROWTH

It can be done by:
1) Capping the main NADH re-oxidation rxn (NADH16pp)
2) Adding a drain reaction for NADH. This reaction will create a flux that demands NADH without regenerating NAD⁺.

```python
from cobra.io import load_model

# load
model = load_model("iML1515")

solution = model.optimize()
print('og solution:')
print(solution.objective_value)
print()

#1- Capping the main NADH re-oxidation rxn (NADH16pp)
nadh_dehyd = model.reactions.get_by_id("NADH16pp")  # NADH dehydrogenase I
nadh_dehyd.upper_bound = 3

#2- Adding a drain reaction for NADH
nadh = model.metabolites.nadh_c
dm_nadh = cobra.Reaction("DM_nadh_c")
dm_nadh.add_metabolites({nadh: -1})       # drains 1 NADH → nothing
dm_nadh.lower_bound = dm_nadh.upper_bound = 0.1
model.add_reactions([dm_nadh])


solution = model.optimize()
print('NEW solution:')
print(solution.objective_value)
print()
```

```python
import cobra
from cobra.io import load_model
from typing import Tuple, Dict, Any


def select_ecomics_condition(strain: str, medium_id: str, medium: str, stress: str) -> Tuple[cobra.Model, str]:
    """
    Modify iML1515 GEM based on specified experimental conditions.
    
    Parameters:
    -----------
    strain : str
        E. coli strain type. Options: 'W3110', 'MG1655', 'BW25113'
    medium_id : str
        Medium identifier. Options: 'MD066', 'MD120', 'MD004', 'MD121'
    medium : str
        Medium description. Options: 'synthetic+Glu', 'synthetic+Glu+KNO3', 'MOPS+Glu(0.4%)', 'M9+Glu'
    stress : str
        Stress condition. Options: 'none', 'NADH-limitation', 'ATP-limitation'
    
    Returns:
    --------
    Tuple[cobra.Model, str]
        Modified GEM model and condition string
    
    Raises:
    -------
    ValueError
        If any input parameter is not in the allowed options
    """
    
    # Validate input parameters
    valid_strains = {'W3110', 'MG1655', 'BW25113'}
    valid_medium_ids = {'MD066', 'MD120', 'MD004', 'MD121'}
    valid_mediums = {'synthetic+Glu', 'synthetic+Glu+KNO3', 'MOPS+Glu(0.4%)', 'M9+Glu'}
    valid_stress = {'none', 'NADH-limitation', 'ATP-limitation'}
    
    if strain not in valid_strains:
        raise ValueError(f"Invalid strain '{strain}'. Must be one of {valid_strains}")
    if medium_id not in valid_medium_ids:
        raise ValueError(f"Invalid medium_id '{medium_id}'. Must be one of {valid_medium_ids}")
    if medium not in valid_mediums:
        raise ValueError(f"Invalid medium '{medium}'. Must be one of {valid_mediums}")
    if stress not in valid_stress:
        raise ValueError(f"Invalid stress '{stress}'. Must be one of {valid_stress}")
    
    # Load the iML1515 model (assuming it's available via cobra)
    model = cobra.io.load_model("iML1515")
    
    # Apply medium-specific modifications
    _apply_medium_conditions(model, medium_id)
    
    # Apply stress-specific modifications
    _apply_stress_conditions(model, stress)
    
    # Create condition string
    condition_string = f"Strain:{strain}_MediumID:{medium_id}_Medium:{medium}_Stress:{stress}"
    
    return model, condition_string


def _apply_medium_conditions(model: cobra.Model, medium_id: str) -> None:
    """
    Apply medium-specific modifications to the model.
    
    Parameters:
    -----------
    model : cobra.Model
        The genome-scale model to modify
    medium_id : str
        Medium identifier
    """
    
    if medium_id in ['MD066', 'MD120', 'MD121']:
        # Define custom synthetic+Glu medium
        medium = {
            "EX_glc__D_e": -10
        }
        
    elif medium_id == 'MD004':
        # Define custom synthetic+Glu medium with higher glucose uptake
        medium = {
            "EX_glc__D_e": -10
        }
    
    # Apply the medium by setting lower bounds
    for rxn_id, uptake in medium.items():
        try:
            rxn = model.reactions.get_by_id(rxn_id)
            rxn.lower_bound = uptake
            print(f"Set {rxn_id} lower bound to {uptake}")
        except KeyError:
            print(f"Reaction {rxn_id} not found in model.")


def _apply_stress_conditions(model: cobra.Model, stress: str) -> None:
    """
    Apply stress-specific modifications to the model.
    
    Parameters:
    -----------
    model : cobra.Model
        The genome-scale model to modify
    stress : str
        Stress condition
    """
    
    if stress == 'NADH-limitation':
        try:
            # Limit NADH dehydrogenase I
            nadh_dehyd = model.reactions.get_by_id("NADH16pp")
            nadh_dehyd.upper_bound = 3
            print("Applied NADH dehydrogenase limitation")
            
            # Add NADH drain reaction
            nadh = model.metabolites.nadh_c
            dm_nadh = cobra.Reaction("DM_nadh_c")
            dm_nadh.name = "NADH drain reaction"
            dm_nadh.add_metabolites({nadh: -1})  # drains 1 NADH → nothing
            dm_nadh.lower_bound = dm_nadh.upper_bound = 0.1
            model.add_reactions([dm_nadh])
            print("Added NADH drain reaction")
            
        except KeyError as e:
            print(f"Error applying NADH limitation: {e}")
    
    elif stress == 'ATP-limitation':
        try:
            # Increase ATP maintenance requirement
            atp_maint = model.reactions.get_by_id("ATPM")
            atp_maint.lower_bound = 20  # mmol · gDW⁻¹ · h⁻¹
            atp_maint.upper_bound = 20
            print("Increased ATP maintenance requirement")
            
            # Limit ATP synthase
            atp_synth = model.reactions.get_by_id("ATPS4rpp")
            atp_synth.upper_bound = 5
            print("Limited ATP synthase capacity")
            
        except KeyError as e:
            print(f"Error applying ATP limitation: {e}")
    
    elif stress == 'none':
        # No stress modifications needed
        pass
```

```python
import pandas as pd
import cobra
from cobra import flux_analysis
import numpy as np
from typing import Dict, Any, Optional
import warnings


def compare_ecomics_fluxomics(csv_file: str, modified_gem: cobra.Model, conditions: str) -> pd.DataFrame:
    """
    Compare experimental fluxomics data with FBA and pFBA simulation results.
    
    Parameters:
    -----------
    csv_file : str
        Path to the 'fluxomics_ecomics_mapped.csv' file
    modified_gem : cobra.Model
        Modified GEM from select_ecomics_condition module
    conditions : str
        Condition string in format "Strain:X_MediumID:Y_Medium:Z_Stress:W"
    
    Returns:
    --------
    pd.DataFrame
        DataFrame with columns: 'rxn_id', 'fluxomics_exp_{STRAIN}_{MEDIUM}_{STRESS}', 'FBA', 'pFBA'
    
    Raises:
    -------
    FileNotFoundError
        If the CSV file is not found
    ValueError
        If conditions don't match any row in the experimental data
    """
    
    # Load experimental fluxomics data
    try:
        fluxomics_df = pd.read_csv(csv_file)
        print(f"Loaded fluxomics data with {len(fluxomics_df)} rows and {len(fluxomics_df.columns)} columns")
    except FileNotFoundError:
        raise FileNotFoundError(f"Could not find fluxomics file: {csv_file}")
    except Exception as e:
        raise Exception(f"Error loading fluxomics data: {e}")
    
    # Parse conditions string
    strain, medium_id, medium, stress = _parse_conditions(conditions)
    print(f"Looking for conditions: Strain={strain}, MediumID={medium_id}, Medium={medium}, Stress={stress}")
    
    # Show possible condition combinations
    _show_available_conditions(fluxomics_df)
    
    # Find matching experimental data
    exp_data = _get_experimental_fluxes(fluxomics_df, strain, medium_id, medium, stress)
    
    # Run FBA and pFBA simulations
    fba_fluxes, pfba_fluxes = _run_simulations(modified_gem)
    
    # Create comparison dataframe
    comparison_df = _create_comparison_dataframe(exp_data, fba_fluxes, pfba_fluxes, strain, medium, stress)
    
    return comparison_df


def _parse_conditions(conditions: str) -> tuple:
    """
    Parse condition string into individual components.
    
    Parameters:
    -----------
    conditions : str
        Condition string in format "Strain:X_MediumID:Y_Medium:Z_Stress:W"
    
    Returns:
    --------
    tuple
        (strain, medium_id, medium, stress)
    """
    try:
        parts = conditions.split('_')
        strain = parts[0].split(':')[1]
        medium_id = parts[1].split(':')[1]
        medium = parts[2].split(':')[1]
        stress = parts[3].split(':')[1]
        
        return strain, medium_id, medium, stress
    except (IndexError, ValueError) as e:
        raise ValueError(f"Invalid conditions format: {conditions}. Expected format: 'Strain:X_MediumID:Y_Medium:Z_Stress:W'")


# Using first flux row
def _get_experimental_fluxes(fluxomics_df: pd.DataFrame, strain: str, medium_id: str, medium: str, stress: str) -> Dict[str, float]:
    """
    Extract experimental flux values for specified conditions.
    
    Parameters:
    -----------
    fluxomics_df : pd.DataFrame
        Experimental fluxomics data
    strain : str
        E. coli strain
    medium_id : str
        Medium identifier
    medium : str
        Medium description
    stress : str
        Stress condition
    
    Returns:
    --------
    Dict[str, float]
        Dictionary mapping reaction IDs to experimental flux values
    """
    # Clean up string values to handle whitespace and case issues
    fluxomics_df_clean = fluxomics_df.copy()
    
    # Clean string columns
    string_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    for col in string_cols:
        if col in fluxomics_df_clean.columns:
            fluxomics_df_clean[col] = fluxomics_df_clean[col].astype(str).str.strip()
    
    # Clean input parameters
    strain = str(strain).strip()
    medium_id = str(medium_id).strip()
    medium = str(medium).strip()
    stress = str(stress).strip()
    
    # Try exact matching first
    mask = (
        (fluxomics_df_clean['Strain'] == strain) &
        (fluxomics_df_clean['MediumID'] == medium_id) &
        (fluxomics_df_clean['Medium'] == medium) &
        (fluxomics_df_clean['Stress'] == stress)
    )
    
    matching_rows = fluxomics_df_clean[mask]
    
    # If no exact match, try case-insensitive matching
    if len(matching_rows) == 0:
        print("Exact match failed, trying case-insensitive matching...")
        mask = (
            (fluxomics_df_clean['Strain'].str.lower() == strain.lower()) &
            (fluxomics_df_clean['MediumID'].str.lower() == medium_id.lower()) &
            (fluxomics_df_clean['Medium'].str.lower() == medium.lower()) &
            (fluxomics_df_clean['Stress'].str.lower() == stress.lower())
        )
        matching_rows = fluxomics_df_clean[mask]
    
    # If still no match, try partial matching for medium
    if len(matching_rows) == 0:
        print("Case-insensitive match failed, trying partial medium matching...")
        # Extract key parts of medium for matching
        medium_key = medium.lower().replace(' ', '').replace('(', '').replace(')', '').replace('%', '')
        
        medium_matches = []
        for idx, row in fluxomics_df_clean.iterrows():
            row_medium = str(row['Medium']).lower().replace(' ', '').replace('(', '').replace(')', '').replace('%', '')
            if (row['Strain'].lower() == strain.lower() and 
                row['MediumID'].lower() == medium_id.lower() and
                row['Stress'].lower() == stress.lower() and
                medium_key in row_medium):
                medium_matches.append(idx)
        
        if medium_matches:
            matching_rows = fluxomics_df_clean.loc[medium_matches]
    
    if len(matching_rows) == 0:
        print("\nDEBUG: No matches found. Here's what we're looking for vs what's available:")
        print(f"Target: Strain='{strain}', MediumID='{medium_id}', Medium='{medium}', Stress='{stress}'")
        print("\nAvailable combinations:")
        condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
        unique_combinations = fluxomics_df_clean[condition_cols].drop_duplicates()
        for idx, row in unique_combinations.iterrows():
            print(f"  Strain='{row['Strain']}', MediumID='{row['MediumID']}', Medium='{row['Medium']}', Stress='{row['Stress']}'")
        
        raise ValueError(f"No experimental data found for conditions: Strain={strain}, MediumID={medium_id}, Medium={medium}, Stress={stress}")
    
    if len(matching_rows) > 1:
        warnings.warn(f"Multiple rows ({len(matching_rows)}) found for conditions. Using the first one.")
    
    # Extract flux values (excluding condition columns)
    exp_row = matching_rows.iloc[0]
    condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    flux_cols = [col for col in fluxomics_df_clean.columns if col not in condition_cols]
    
    exp_fluxes = {}
    for rxn_id in flux_cols:
        flux_value = exp_row[rxn_id]
        if pd.notna(flux_value):  # Only include non-NaN values
            try:
                exp_fluxes[rxn_id] = float(flux_value)
            except (ValueError, TypeError):
                print(f"Warning: Could not convert flux value for {rxn_id}: {flux_value}")
    
    print(f"Successfully extracted {len(exp_fluxes)} experimental flux values")
    return exp_fluxes


def _run_simulations(model: cobra.Model) -> tuple:
    """
    Run FBA and pFBA simulations on the modified model.
    
    Parameters:
    -----------
    model : cobra.Model
        Modified genome-scale model
    
    Returns:
    --------
    tuple
        (fba_fluxes, pfba_fluxes) - dictionaries mapping reaction IDs to flux values
    """
    try:
        # Run FBA
        print("Running FBA simulation...")
        fba_solution = model.optimize()
        
        if fba_solution.status != 'optimal':
            warnings.warn(f"FBA optimization status: {fba_solution.status}")
        
        fba_fluxes = dict(fba_solution.fluxes)
        print(f"FBA objective value: {fba_solution.objective_value:.4f}")
        
        # Run pFBA
        print("Running pFBA simulation...")
        pfba_solution = flux_analysis.pfba(model)
        
        if pfba_solution.status != 'optimal':
            warnings.warn(f"pFBA optimization status: {pfba_solution.status}")
        
        pfba_fluxes = dict(pfba_solution.fluxes)
        print(f"pFBA objective value: {pfba_solution.objective_value:.4f}")
        
        return fba_fluxes, pfba_fluxes
        
    except Exception as e:
        raise Exception(f"Error running simulations: {e}")


def _create_comparison_dataframe(exp_data: Dict[str, float], fba_fluxes: Dict[str, float], 
                                pfba_fluxes: Dict[str, float], strain: str, medium: str, stress: str) -> pd.DataFrame:
    """
    Create comparison dataframe with experimental and simulation results.
    
    Parameters:
    -----------
    exp_data : Dict[str, float]
        Experimental flux data
    fba_fluxes : Dict[str, float]
        FBA simulation results
    pfba_fluxes : Dict[str, float]
        pFBA simulation results
    strain : str
        E. coli strain
    medium : str
        Medium description
    stress : str
        Stress condition
    
    Returns:
    --------
    pd.DataFrame
        Comparison dataframe
    """
    # Get all unique reaction IDs
    all_rxn_ids = set(exp_data.keys()) | set(fba_fluxes.keys()) | set(pfba_fluxes.keys())
    
    # Create experimental column name
    exp_col_name = f"fluxomics_exp_{strain}_{medium}_{stress}"
    
    # Build comparison data
    comparison_data = []
    
    for rxn_id in sorted(all_rxn_ids):
        row = {
            'rxn_id': rxn_id,
            exp_col_name: exp_data.get(rxn_id, np.nan),
            'FBA': fba_fluxes.get(rxn_id, np.nan),
            'pFBA': pfba_fluxes.get(rxn_id, np.nan)
        }
        comparison_data.append(row)
    
    comparison_df = pd.DataFrame(comparison_data)
    
    # Filter out rows where all simulation values are NaN or zero
    mask = ~(
        (comparison_df['FBA'].isna() | (comparison_df['FBA'] == 0)) &
        (comparison_df['pFBA'].isna() | (comparison_df['pFBA'] == 0)) &
        comparison_df[exp_col_name].isna()
    )
    
    comparison_df = comparison_df[mask].reset_index(drop=True)
    
    print(f"Created comparison dataframe with {len(comparison_df)} reactions")
    return comparison_df


def _show_available_conditions(fluxomics_df: pd.DataFrame) -> None:
    """
    Display available experimental conditions for debugging.
    
    Parameters:
    -----------
    fluxomics_df : pd.DataFrame
        Experimental fluxomics data
    """
    print("\n=== AVAILABLE EXPERIMENTAL CONDITIONS ===")
    condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    
    for col in condition_cols:
        if col in fluxomics_df.columns:
            unique_vals = fluxomics_df[col].unique()
            print(f"{col}: {list(unique_vals)}")
    
    print(f"\nTotal number of experimental conditions: {len(fluxomics_df)}")
    print("=" * 45)


def inspect_fluxomics_data(csv_file: str) -> pd.DataFrame:
    """
    Utility function to inspect the fluxomics data structure.
    
    Parameters:
    -----------
    csv_file : str
        Path to the fluxomics CSV file
    
    Returns:
    --------
    pd.DataFrame
        Summary of available conditions
    """
    try:
        df = pd.read_csv(csv_file)
        print(f"Dataset shape: {df.shape}")
        print(f"Columns: {list(df.columns)}")
        
        # Show available conditions
        _show_available_conditions(df)
        
        # Show first few rows
        print("\nFirst 3 rows:")
        print(df.head(3))
        
        # Return unique condition combinations
        condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
        available_cols = [col for col in condition_cols if col in df.columns]
        
        if available_cols:
            unique_conditions = df[available_cols].drop_duplicates().reset_index(drop=True)
            print(f"\nUnique condition combinations ({len(unique_conditions)}):")
            print(unique_conditions)
            return unique_conditions
        else:
            print("Warning: Standard condition columns not found!")
            return df
            
    except Exception as e:
        print(f"Error inspecting data: {e}")
        return pd.DataFrame()


```

```python
def order_by_data_completeness(fluxomics_mapped_df: pd.DataFrame, start_col_index: int = 4) -> pd.DataFrame:
    """
    Order rows in the fluxomics dataframe by data completeness (least zero values first).
    
    Parameters:
        fluxomics_mapped_df : pd.DataFrame
            The fluxomics dataframe to be ordered
        start_col_index : int, default=4
            The column index from which to start counting zeros (0-based indexing)
    
    Returns:
    pd.DataFrame
        Dataframe ordered by data completeness (rows with least zeros first)
    """
    # Create a copy to avoid modifying the original dataframe
    df_copy = fluxomics_mapped_df.copy()
    
    # Get the columns from start_col_index onwards
    flux_columns = df_copy.columns[start_col_index:]
    
    # Count zeros in each row for the specified columns
    # This handles both actual zeros and NaN values (treating NaN as missing data)
    zero_counts = []
    
    for idx, row in df_copy.iterrows():
        # Count zeros and NaN values in flux columns
        flux_values = row[flux_columns]
        zero_count = (flux_values == 0).sum() + flux_values.isna().sum()
        zero_counts.append(zero_count)
    
    # Add the zero count as a temporary column for sorting
    df_copy['_zero_count'] = zero_counts
    
    # Add original index as a column to maintain stable sort
    df_copy['_original_index'] = df_copy.index
    
    # Sort by zero count (ascending - least zeros first), then by original index to maintain stable sort
    df_sorted = df_copy.sort_values(['_zero_count', '_original_index'], ascending=[True, True])
    
    # Remove the temporary columns
    df_sorted = df_sorted.drop(['_zero_count', '_original_index'], axis=1)
    
    # Reset index to have clean sequential numbering
    df_sorted = df_sorted.reset_index(drop=True)
    
    print(f"Ordered {len(df_sorted)} rows by data completeness")
    print(f"Top row has {min(zero_counts)} zero/missing values")
    print(f"Bottom row has {max(zero_counts)} zero/missing values")
    
    return df_sorted


def analyze_data_completeness(fluxomics_mapped_df: pd.DataFrame, start_col_index: int = 4) -> pd.DataFrame:
    """
    Analyze and display data completeness statistics for the fluxomics dataframe.
    
    Parameters:
    -----------
    fluxomics_mapped_df : pd.DataFrame
        The fluxomics dataframe to analyze
    start_col_index : int, default=4
        The column index from which to start counting zeros (0-based indexing)
    
    Returns:
    --------
    pd.DataFrame
        Summary statistics of data completeness per row
    """
    # Get the columns from start_col_index onwards
    flux_columns = fluxomics_mapped_df.columns[start_col_index:]
    total_flux_cols = len(flux_columns)
    
    # Calculate completeness statistics for each row
    completeness_stats = []
    
    for idx, row in fluxomics_mapped_df.iterrows():
        flux_values = row[flux_columns]
        
        # Count different types of values
        zero_count = (flux_values == 0).sum()
        nan_count = flux_values.isna().sum()
        non_zero_count = ((flux_values != 0) & flux_values.notna()).sum()
        
        # Calculate percentages
        zero_pct = (zero_count / total_flux_cols) * 100
        nan_pct = (nan_count / total_flux_cols) * 100
        non_zero_pct = (non_zero_count / total_flux_cols) * 100
        
        # Create condition identifier from first few columns
        condition_cols = fluxomics_mapped_df.columns[:start_col_index]
        condition_id = "_".join([str(row[col]) for col in condition_cols])
        
        completeness_stats.append({
            'row_index': idx,
            'condition': condition_id,
            'total_flux_cols': total_flux_cols,
            'zero_count': zero_count,
            'nan_count': nan_count,
            'non_zero_count': non_zero_count,
            'missing_total': zero_count + nan_count,
            'zero_pct': zero_pct,
            'nan_pct': nan_pct,
            'non_zero_pct': non_zero_pct,
            'data_completeness_pct': non_zero_pct
        })
    
    completeness_df = pd.DataFrame(completeness_stats)
    
    # Sort by data completeness (highest first)
    completeness_df = completeness_df.sort_values('data_completeness_pct', ascending=False)
    
    print(f"Data completeness analysis for {len(flux_columns)} flux columns:")
    print(f"Best row: {completeness_df.iloc[0]['data_completeness_pct']:.1f}% complete")
    print(f"Worst row: {completeness_df.iloc[-1]['data_completeness_pct']:.1f}% complete")
    print(f"Average completeness: {completeness_df['data_completeness_pct'].mean():.1f}%")
    
    return completeness_df


def get_top_complete_conditions(fluxomics_mapped_df: pd.DataFrame, top_n: int = 10, start_col_index: int = 4) -> pd.DataFrame:
    """
    Get the top N conditions with the most complete flux data.
    
    Parameters:
    -----------
    fluxomics_mapped_df : pd.DataFrame
        The fluxomics dataframe to analyze
    top_n : int, default=10
        Number of top conditions to return
    start_col_index : int, default=4
        The column index from which to start counting zeros (0-based indexing)
    
    Returns:
    --------
    pd.DataFrame
        Top N rows with most complete data
    """
    # Order by data completeness
    ordered_df = order_by_data_completeness(fluxomics_mapped_df, start_col_index)
    
    # Get top N rows
    top_conditions = ordered_df.head(top_n)
    
    print(f"Top {min(top_n, len(ordered_df))} conditions with most complete flux data:")
    
    # Display summary for each top condition
    flux_columns = ordered_df.columns[start_col_index:]
    for i, (idx, row) in enumerate(top_conditions.iterrows()):
        flux_values = row[flux_columns]
        non_zero_count = ((flux_values != 0) & flux_values.notna()).sum()
        completeness_pct = (non_zero_count / len(flux_columns)) * 100
        
        condition_cols = ordered_df.columns[:start_col_index]
        condition_id = "_".join([str(row[col]) for col in condition_cols])
        
        print(f"  {i+1}. {condition_id}: {completeness_pct:.1f}% complete ({non_zero_count}/{len(flux_columns)} flux values)")
    
    return top_conditions
```

```python
import pandas as pd
import cobra
from cobra import flux_analysis
import numpy as np
from typing import Dict, Any, Optional
import warnings


def compare_ecomics_fluxomics(csv_file: str, modified_gem: cobra.Model, conditions: str) -> pd.DataFrame:
    """
    Compare experimental fluxomics data with FBA and pFBA simulation results.
    
    Parameters:
    -----------
    csv_file : str
        Path to the 'fluxomics_ecomics_mapped.csv' file
    modified_gem : cobra.Model
        Modified GEM from select_ecomics_condition module
    conditions : str
        Condition string in format "Strain:X_MediumID:Y_Medium:Z_Stress:W"
    
    Returns:
    --------
    pd.DataFrame
        DataFrame with columns: 'rxn_id', 'fluxomics_exp_{STRAIN}_{MEDIUM}_{STRESS}', 'FBA', 'pFBA'
    
    Raises:
    -------
    FileNotFoundError
        If the CSV file is not found
    ValueError
        If conditions don't match any row in the experimental data
    """
    
    # Load experimental fluxomics data
    try:
        fluxomics_df = pd.read_csv(csv_file)
        print(f"Loaded fluxomics data with {len(fluxomics_df)} rows and {len(fluxomics_df.columns)} columns")
    except FileNotFoundError:
        raise FileNotFoundError(f"Could not find fluxomics file: {csv_file}")
    except Exception as e:
        raise Exception(f"Error loading fluxomics data: {e}")
    
    # Parse conditions string
    strain, medium_id, medium, stress = _parse_conditions(conditions)
    print(f"Looking for conditions: Strain={strain}, MediumID={medium_id}, Medium={medium}, Stress={stress}")
    
    # Show possible condition combinations
    _show_available_conditions(fluxomics_df)
    
    # Find matching experimental data
    exp_data, flux_stats_df = _get_experimental_fluxes(fluxomics_df, strain, medium_id, medium, stress)
    
    # Run FBA and pFBA simulations
    fba_fluxes, pfba_fluxes = _run_simulations(modified_gem)
    
    # Create comparison dataframe
    comparison_df = _create_comparison_dataframe(exp_data, fba_fluxes, pfba_fluxes, strain, medium, stress)
    
    return comparison_df, flux_stats_df


def _get_experimental_fluxes(fluxomics_df: pd.DataFrame, strain: str, medium_id: str, medium: str, stress: str) -> Tuple[Dict[str, float], pd.DataFrame]:
    """
    Extract experimental flux values for specified conditions.
    
    Parameters:
    -----------
    fluxomics_df : pd.DataFrame
        Experimental fluxomics data
    strain : str
        E. coli strain
    medium_id : str
        Medium identifier
    medium : str
        Medium description
    stress : str
        Stress condition
    
    Returns:
    --------
    Tuple[Dict[str, float], pd.DataFrame]
        - Dictionary mapping reaction IDs to experimental flux values (averaged if multiple rows match)
        - DataFrame with individual flux values, average, and standard deviation for each reaction
    """
    # Clean up string values to handle whitespace and case issues
    fluxomics_df_clean = fluxomics_df.copy()
    
    # Clean string columns
    string_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    for col in string_cols:
        if col in fluxomics_df_clean.columns:
            fluxomics_df_clean[col] = fluxomics_df_clean[col].astype(str).str.strip()
    
    # Clean input parameters
    strain = str(strain).strip()
    medium_id = str(medium_id).strip()
    medium = str(medium).strip()
    stress = str(stress).strip()
    
    # Try exact matching first
    mask = (
        (fluxomics_df_clean['Strain'] == strain) &
        (fluxomics_df_clean['MediumID'] == medium_id) &
        (fluxomics_df_clean['Medium'] == medium) &
        (fluxomics_df_clean['Stress'] == stress)
    )
    
    matching_rows = fluxomics_df_clean[mask]
    
    # If no exact match, try case-insensitive matching
    if len(matching_rows) == 0:
        print("Exact match failed, trying case-insensitive matching...")
        mask = (
            (fluxomics_df_clean['Strain'].str.lower() == strain.lower()) &
            (fluxomics_df_clean['MediumID'].str.lower() == medium_id.lower()) &
            (fluxomics_df_clean['Medium'].str.lower() == medium.lower()) &
            (fluxomics_df_clean['Stress'].str.lower() == stress.lower())
        )
        matching_rows = fluxomics_df_clean[mask]
    
    # If still no match, try partial matching for medium
    if len(matching_rows) == 0:
        print("Case-insensitive match failed, trying partial medium matching...")
        # Extract key parts of medium for matching
        medium_key = medium.lower().replace(' ', '').replace('(', '').replace(')', '').replace('%', '')
        
        medium_matches = []
        for idx, row in fluxomics_df_clean.iterrows():
            row_medium = str(row['Medium']).lower().replace(' ', '').replace('(', '').replace(')', '').replace('%', '')
            if (row['Strain'].lower() == strain.lower() and 
                row['MediumID'].lower() == medium_id.lower() and
                row['Stress'].lower() == stress.lower() and
                medium_key in row_medium):
                medium_matches.append(idx)
        
        if medium_matches:
            matching_rows = fluxomics_df_clean.loc[medium_matches]
    
    if len(matching_rows) == 0:
        print("\nDEBUG: No matches found. Here's what we're looking for vs what's available:")
        print(f"Target: Strain='{strain}', MediumID='{medium_id}', Medium='{medium}', Stress='{stress}'")
        print("\nAvailable combinations:")
        condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
        unique_combinations = fluxomics_df_clean[condition_cols].drop_duplicates()
        for idx, row in unique_combinations.iterrows():
            print(f"  Strain='{row['Strain']}', MediumID='{row['MediumID']}', Medium='{row['Medium']}', Stress='{row['Stress']}'")
        
        raise ValueError(f"No experimental data found for conditions: Strain={strain}, MediumID={medium_id}, Medium={medium}, Stress={stress}")
    
    # Handle multiple matching rows by including all of them
    if len(matching_rows) > 1:
        print(f"Multiple rows ({len(matching_rows)}) found for conditions. Including all matches and calculating statistics.")
    
    # Extract flux values from all matching rows
    condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    flux_cols = [col for col in fluxomics_df_clean.columns if col not in condition_cols]
    
    # Create dataframe to store all individual flux values with statistics
    flux_stats_data = []
    exp_fluxes = {}
    
    # Process each reaction
    for rxn_id in flux_cols:
        row_data = {'rxn_id': rxn_id}
        reaction_values = []
        
        # Collect flux values from all matching rows
        for i, (idx, exp_row) in enumerate(matching_rows.iterrows()):
            flux_value = exp_row[rxn_id]
            col_name = f"exp_{i+1}"
            
            if pd.notna(flux_value):
                try:
                    float_value = float(flux_value)
                    row_data[col_name] = float_value
                    reaction_values.append(float_value)
                except (ValueError, TypeError):
                    print(f"Warning: Could not convert flux value for {rxn_id}: {flux_value}")
                    row_data[col_name] = np.nan
            else:
                row_data[col_name] = np.nan
        
        # Calculate statistics
        if reaction_values:  # Only if we have valid values
            row_data['average'] = np.mean(reaction_values)
            row_data['std_dev'] = np.std(reaction_values, ddof=1) if len(reaction_values) > 1 else 0.0
            row_data['count'] = len(reaction_values)
            
            # Add to the main experimental fluxes dictionary
            exp_fluxes[rxn_id] = row_data['average']
            
            if len(reaction_values) > 1:
                print(f"  Reaction {rxn_id}: averaged {len(reaction_values)} values -> {row_data['average']:.4f} ± {row_data['std_dev']:.4f}")
        else:
            row_data['average'] = np.nan
            row_data['std_dev'] = np.nan
            row_data['count'] = 0
        
        flux_stats_data.append(row_data)
    
    # Create the detailed statistics dataframe
    flux_stats_df = pd.DataFrame(flux_stats_data)
    
    # Remove reactions that have no valid measurements
    flux_stats_df = flux_stats_df[flux_stats_df['count'] > 0].reset_index(drop=True)
    
    print(f"Successfully extracted {len(exp_fluxes)} experimental flux values from {len(matching_rows)} matching row(s)")
    print(f"Created detailed flux statistics dataframe with {len(flux_stats_df)} reactions")
    
    return exp_fluxes, flux_stats_df


def _parse_conditions(conditions: str) -> tuple:
    """
    Parse condition string into individual components.
    
    Parameters:
    -----------
    conditions : str
        Condition string in format "Strain:X_MediumID:Y_Medium:Z_Stress:W"
    
    Returns:
    --------
    tuple
        (strain, medium_id, medium, stress)
    """
    try:
        parts = conditions.split('_')
        strain = parts[0].split(':')[1]
        medium_id = parts[1].split(':')[1]
        medium = parts[2].split(':')[1]
        stress = parts[3].split(':')[1]
        
        return strain, medium_id, medium, stress
    except (IndexError, ValueError) as e:
        raise ValueError(f"Invalid conditions format: {conditions}. Expected format: 'Strain:X_MediumID:Y_Medium:Z_Stress:W'")


# Using first flux row
def _get_experimental_fluxes(fluxomics_df: pd.DataFrame, strain: str, medium_id: str, medium: str, stress: str) -> Dict[str, float]:
    """
    Extract experimental flux values for specified conditions.
    
    Parameters:
    -----------
    fluxomics_df : pd.DataFrame
        Experimental fluxomics data
    strain : str
        E. coli strain
    medium_id : str
        Medium identifier
    medium : str
        Medium description
    stress : str
        Stress condition
    
    Returns:
    --------
    Dict[str, float]
        Dictionary mapping reaction IDs to experimental flux values
    """
    # Clean up string values to handle whitespace and case issues
    fluxomics_df_clean = fluxomics_df.copy()
    
    # Clean string columns
    string_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    for col in string_cols:
        if col in fluxomics_df_clean.columns:
            fluxomics_df_clean[col] = fluxomics_df_clean[col].astype(str).str.strip()
    
    # Clean input parameters
    strain = str(strain).strip()
    medium_id = str(medium_id).strip()
    medium = str(medium).strip()
    stress = str(stress).strip()
    
    # Try exact matching first
    mask = (
        (fluxomics_df_clean['Strain'] == strain) &
        (fluxomics_df_clean['MediumID'] == medium_id) &
        (fluxomics_df_clean['Medium'] == medium) &
        (fluxomics_df_clean['Stress'] == stress)
    )
    
    matching_rows = fluxomics_df_clean[mask]
    
    # If no exact match, try case-insensitive matching
    if len(matching_rows) == 0:
        print("Exact match failed, trying case-insensitive matching...")
        mask = (
            (fluxomics_df_clean['Strain'].str.lower() == strain.lower()) &
            (fluxomics_df_clean['MediumID'].str.lower() == medium_id.lower()) &
            (fluxomics_df_clean['Medium'].str.lower() == medium.lower()) &
            (fluxomics_df_clean['Stress'].str.lower() == stress.lower())
        )
        matching_rows = fluxomics_df_clean[mask]
    
    # If still no match, try partial matching for medium
    if len(matching_rows) == 0:
        print("Case-insensitive match failed, trying partial medium matching...")
        # Extract key parts of medium for matching
        medium_key = medium.lower().replace(' ', '').replace('(', '').replace(')', '').replace('%', '')
        
        medium_matches = []
        for idx, row in fluxomics_df_clean.iterrows():
            row_medium = str(row['Medium']).lower().replace(' ', '').replace('(', '').replace(')', '').replace('%', '')
            if (row['Strain'].lower() == strain.lower() and 
                row['MediumID'].lower() == medium_id.lower() and
                row['Stress'].lower() == stress.lower() and
                medium_key in row_medium):
                medium_matches.append(idx)
        
        if medium_matches:
            matching_rows = fluxomics_df_clean.loc[medium_matches]
    
    if len(matching_rows) == 0:
        print("\nDEBUG: No matches found. Here's what we're looking for vs what's available:")
        print(f"Target: Strain='{strain}', MediumID='{medium_id}', Medium='{medium}', Stress='{stress}'")
        print("\nAvailable combinations:")
        condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
        unique_combinations = fluxomics_df_clean[condition_cols].drop_duplicates()
        for idx, row in unique_combinations.iterrows():
            print(f"  Strain='{row['Strain']}', MediumID='{row['MediumID']}', Medium='{row['Medium']}', Stress='{row['Stress']}'")
        
        raise ValueError(f"No experimental data found for conditions: Strain={strain}, MediumID={medium_id}, Medium={medium}, Stress={stress}")
    
    if len(matching_rows) > 1:
        warnings.warn(f"Multiple rows ({len(matching_rows)}) found for conditions. Using the first one.")
    
    # Extract flux values (excluding condition columns)
    exp_row = matching_rows.iloc[0]
    condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    flux_cols = [col for col in fluxomics_df_clean.columns if col not in condition_cols]
    
    exp_fluxes = {}
    for rxn_id in flux_cols:
        flux_value = exp_row[rxn_id]
        if pd.notna(flux_value):  # Only include non-NaN values
            try:
                exp_fluxes[rxn_id] = float(flux_value)
            except (ValueError, TypeError):
                print(f"Warning: Could not convert flux value for {rxn_id}: {flux_value}")
    
    print(f"Successfully extracted {len(exp_fluxes)} experimental flux values")
    return exp_fluxes, flux_stats_df




def _run_simulations(model: cobra.Model) -> tuple:
    """
    Run FBA and pFBA simulations on the modified model.
    
    Parameters:
    -----------
    model : cobra.Model
        Modified genome-scale model
    
    Returns:
    --------
    tuple
        (fba_fluxes, pfba_fluxes) - dictionaries mapping reaction IDs to flux values
    """
    try:
        # Run FBA
        print("Running FBA simulation...")
        fba_solution = model.optimize()
        
        if fba_solution.status != 'optimal':
            warnings.warn(f"FBA optimization status: {fba_solution.status}")
        
        fba_fluxes = dict(fba_solution.fluxes)
        print(f"FBA objective value: {fba_solution.objective_value:.4f}")
        
        # Run pFBA
        print("Running pFBA simulation...")
        pfba_solution = flux_analysis.pfba(model)
        
        if pfba_solution.status != 'optimal':
            warnings.warn(f"pFBA optimization status: {pfba_solution.status}")
        
        pfba_fluxes = dict(pfba_solution.fluxes)
        print(f"pFBA objective value: {pfba_solution.objective_value:.4f}")
        
        return fba_fluxes, pfba_fluxes
        
    except Exception as e:
        raise Exception(f"Error running simulations: {e}")


def _create_comparison_dataframe(exp_data: Dict[str, float], fba_fluxes: Dict[str, float], 
                                pfba_fluxes: Dict[str, float], strain: str, medium: str, stress: str) -> pd.DataFrame:
    """
    Create comparison dataframe with experimental and simulation results.
    
    Parameters:
    -----------
    exp_data : Dict[str, float]
        Experimental flux data
    fba_fluxes : Dict[str, float]
        FBA simulation results
    pfba_fluxes : Dict[str, float]
        pFBA simulation results
    strain : str
        E. coli strain
    medium : str
        Medium description
    stress : str
        Stress condition
    
    Returns:
    --------
    pd.DataFrame
        Comparison dataframe
    """
    # Get all unique reaction IDs
    all_rxn_ids = set(exp_data.keys()) | set(fba_fluxes.keys()) | set(pfba_fluxes.keys())
    
    # Create experimental column name
    exp_col_name = f"fluxomics_exp_{strain}_{medium}_{stress}"
    
    # Build comparison data
    comparison_data = []
    
    for rxn_id in sorted(all_rxn_ids):
        row = {
            'rxn_id': rxn_id,
            exp_col_name: exp_data.get(rxn_id, np.nan),
            'FBA': fba_fluxes.get(rxn_id, np.nan),
            'pFBA': pfba_fluxes.get(rxn_id, np.nan)
        }
        comparison_data.append(row)
    
    comparison_df = pd.DataFrame(comparison_data)
    
    # Filter out rows where all simulation values are NaN or zero
    mask = ~(
        (comparison_df['FBA'].isna() | (comparison_df['FBA'] == 0)) &
        (comparison_df['pFBA'].isna() | (comparison_df['pFBA'] == 0)) &
        comparison_df[exp_col_name].isna()
    )
    
    comparison_df = comparison_df[mask].reset_index(drop=True)
    
    print(f"Created comparison dataframe with {len(comparison_df)} reactions")
    return comparison_df


def _show_available_conditions(fluxomics_df: pd.DataFrame) -> None:
    """
    Display available experimental conditions for debugging.
    
    Parameters:
    -----------
    fluxomics_df : pd.DataFrame
        Experimental fluxomics data
    """
    print("\n=== AVAILABLE EXPERIMENTAL CONDITIONS ===")
    condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
    
    for col in condition_cols:
        if col in fluxomics_df.columns:
            unique_vals = fluxomics_df[col].unique()
            print(f"{col}: {list(unique_vals)}")
    
    print(f"\nTotal number of experimental conditions: {len(fluxomics_df)}")
    print("=" * 45)


def inspect_fluxomics_data(csv_file: str) -> pd.DataFrame:
    """
    Utility function to inspect the fluxomics data structure.
    
    Parameters:
    -----------
    csv_file : str
        Path to the fluxomics CSV file
    
    Returns:
    --------
    pd.DataFrame
        Summary of available conditions
    """
    try:
        df = pd.read_csv(csv_file)
        print(f"Dataset shape: {df.shape}")
        print(f"Columns: {list(df.columns)}")
        
        # Show available conditions
        _show_available_conditions(df)
        
        # Show first few rows
        print("\nFirst 3 rows:")
        print(df.head(3))
        
        # Return unique condition combinations
        condition_cols = ['Strain', 'MediumID', 'Medium', 'Stress']
        available_cols = [col for col in condition_cols if col in df.columns]
        
        if available_cols:
            unique_conditions = df[available_cols].drop_duplicates().reset_index(drop=True)
            print(f"\nUnique condition combinations ({len(unique_conditions)}):")
            print(unique_conditions)
            return unique_conditions
        else:
            print("Warning: Standard condition columns not found!")
            return df
            
    except Exception as e:
        print(f"Error inspecting data: {e}")
        return pd.DataFrame()
    
```
