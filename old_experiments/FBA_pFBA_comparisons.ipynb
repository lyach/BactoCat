## Editable comparison code

```python
import cobra
from cobra.flux_analysis import pfba
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import wilcoxon, ks_2samp

# Load the iML1515 model
model = cobra.io.load_model("iML1515")

solution_fba = model.optimize()
solution_pfba = pfba(model)

df_fluxes = pd.DataFrame({
    'reaction': solution_pfba.fluxes.index,
    'pFBA': solution_pfba.fluxes.values,
    'FBA': solution_fba.fluxes.values
})

# STATISTICAL ASSESSMENT

# Add derived metrics
df_fluxes['abs_pFBA'] = np.abs(df_fluxes['pFBA'])
df_fluxes['abs_FBA'] = np.abs(df_fluxes['FBA'])
df_fluxes['difference'] = df_fluxes['FBA'] - df_fluxes['pFBA']
df_fluxes['abs_difference'] = np.abs(df_fluxes['difference'])
df_fluxes['active_pFBA'] = df_fluxes['abs_pFBA'] > 1e-6
df_fluxes['active_FBA'] = df_fluxes['abs_FBA'] > 1e-6

# 1. BASIC STATISTICS
print("1. BASIC STATISTICS:")
print(f"Total reactions: {len(df_fluxes)}")
print(f"Active reactions (FBA): {df_fluxes['active_FBA'].sum()}")
print(f"Active reactions (pFBA): {df_fluxes['active_pFBA'].sum()}")
print(f"Sum of absolute fluxes (FBA): {df_fluxes['abs_FBA'].sum():.4f}")
print(f"Sum of absolute fluxes (pFBA): {df_fluxes['abs_pFBA'].sum():.4f}")
print(f"Objective value (FBA): {solution_fba.objective_value:.4f}")


# 2. CORRELATION ANALYSIS
print("\n2. CORRELATION ANALYSIS:")
pearson_r, pearson_p = stats.pearsonr(df_fluxes['FBA'], df_fluxes['pFBA'])
spearman_r, spearman_p = stats.spearmanr(df_fluxes['FBA'], df_fluxes['pFBA'])
print(f"Pearson correlation: r = {pearson_r:.4f}, p = {pearson_p:.2e}")
print(f"Spearman correlation: ρ = {spearman_r:.4f}, p = {spearman_p:.2e}")


# 3. FLUX MAGNITUDE COMPARISON
print("\n3. FLUX MAGNITUDE COMPARISON:")
# Wilcoxon signed-rank test for paired samples
# Determines if pFBA significantly reduces flux magnitudes
active_mask = (df_fluxes['active_FBA'] | df_fluxes['active_pFBA'])
active_fluxes = df_fluxes[active_mask]

if len(active_fluxes) > 0:
    wilcoxon_stat, wilcoxon_p = wilcoxon(
        active_fluxes['abs_FBA'], 
        active_fluxes['abs_pFBA'],
        alternative='two-sided'
    )
    print(f"Wilcoxon signed-rank test: W = {wilcoxon_stat:.4f}, p = {wilcoxon_p:.2e}")
    print(f"Interpretation: {'Significantly different' if wilcoxon_p < 0.05 else 'Not significantly different'} flux magnitudes")


# 4. ACTIVE REACTION COMPARISON (McNemar's test)
# Tests if the pattern of active/inactive reactions differs significantly
print("\n4. ACTIVE REACTION COMPARISON:")
# Create contingency table for McNemar's test
both_active = (df_fluxes['active_FBA'] & df_fluxes['active_pFBA']).sum()
fba_only = (df_fluxes['active_FBA'] & ~df_fluxes['active_pFBA']).sum()
pfba_only = (~df_fluxes['active_FBA'] & df_fluxes['active_pFBA']).sum()
both_inactive = (~df_fluxes['active_FBA'] & ~df_fluxes['active_pFBA']).sum()

print(f"Both active: {both_active}")
print(f"FBA only active: {fba_only}")
print(f"pFBA only active: {pfba_only}")
print(f"Both inactive: {both_inactive}")

# McNemar's test (if enough discordant pairs)
if fba_only + pfba_only > 0:
    mcnemar_stat = (abs(fba_only - pfba_only) - 1)**2 / (fba_only + pfba_only)
    mcnemar_p = 1 - stats.chi2.cdf(mcnemar_stat, 1)
    print(f"McNemar's test: χ² = {mcnemar_stat:.4f}, p = {mcnemar_p:.4f}")

# 5. DISTRIBUTION COMPARISON
print("\n5. DISTRIBUTION COMPARISON:")
# KS test on non-zero fluxes
# Compares the overall shape of flux distributions
nonzero_fba = df_fluxes[df_fluxes['active_FBA']]['FBA']
nonzero_pfba = df_fluxes[df_fluxes['active_pFBA']]['pFBA']

if len(nonzero_fba) > 0 and len(nonzero_pfba) > 0:
    ks_stat, ks_p = ks_2samp(nonzero_fba, nonzero_pfba)
    print(f"Kolmogorov-Smirnov test: D = {ks_stat:.4f}, p = {ks_p:.4f}")


# 6. IDENTIFY MOST DIFFERENT REACTIONS
print("\n6. REACTIONS WITH LARGEST DIFFERENCES:")
top_differences = df_fluxes.nlargest(10, 'abs_difference')[['reaction', 'FBA', 'pFBA', 'difference']]
print(top_differences.to_string(index=False))


# VISUALIZATIONS
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('FBA vs pFBA Comparison', fontsize=16)

# 1. Scatter plot of flux values
axes[0,0].scatter(df_fluxes['FBA'], df_fluxes['pFBA'], alpha=0.6, s=20)
axes[0,0].plot([-100, 100], [-100, 100], 'r--', alpha=0.8)
axes[0,0].set_xlabel('FBA Flux')
axes[0,0].set_ylabel('pFBA Flux')
axes[0,0].set_title(f'Flux Correlation\n(r = {pearson_r:.3f})')
axes[0,0].grid(True, alpha=0.3)

# 2. Histogram of flux differences
axes[0,1].hist(df_fluxes['difference'], bins=50, alpha=0.7, edgecolor='black')
axes[0,1].set_xlabel('Flux Difference (FBA - pFBA)')
axes[0,1].set_ylabel('Frequency')
axes[0,1].set_title('Distribution of Flux Differences')
axes[0,1].axvline(0, color='red', linestyle='--', alpha=0.8)

# 3. Box plot of absolute flux magnitudes
flux_data = [df_fluxes[df_fluxes['active_FBA']]['abs_FBA'], 
             df_fluxes[df_fluxes['active_pFBA']]['abs_pFBA']]
axes[0,2].boxplot(flux_data, labels=['FBA', 'pFBA'])
axes[0,2].set_ylabel('Absolute Flux Value')
axes[0,2].set_title('Flux Magnitude Distribution')
axes[0,2].set_yscale('log')

# 4. Active reactions comparison
active_counts = [df_fluxes['active_FBA'].sum(), df_fluxes['active_pFBA'].sum()]
axes[1,0].bar(['FBA', 'pFBA'], active_counts, color=['skyblue', 'lightcoral'])
axes[1,0].set_ylabel('Number of Active Reactions')
axes[1,0].set_title('Active Reactions Count')
for i, v in enumerate(active_counts):
    axes[1,0].text(i, v + 5, str(v), ha='center', va='bottom')

# 5. Cumulative distribution of absolute fluxes
sorted_fba = np.sort(df_fluxes[df_fluxes['active_FBA']]['abs_FBA'])
sorted_pfba = np.sort(df_fluxes[df_fluxes['active_pFBA']]['abs_pFBA'])
axes[1,1].plot(sorted_fba, np.linspace(0, 1, len(sorted_fba)), label='FBA', linewidth=2)
axes[1,1].plot(sorted_pfba, np.linspace(0, 1, len(sorted_pfba)), label='pFBA', linewidth=2)
axes[1,1].set_xlabel('Absolute Flux Value')
axes[1,1].set_ylabel('Cumulative Probability')
axes[1,1].set_title('Cumulative Distribution')
axes[1,1].legend()
axes[1,1].set_xscale('log')

# 6. Heatmap of top different reactions
top_diff_reactions = df_fluxes.nlargest(20, 'abs_difference')
heatmap_data = top_diff_reactions[['FBA', 'pFBA']].T
vmax = np.abs(heatmap_data.values).max()
im = axes[1,2].imshow(heatmap_data, aspect='auto', cmap='RdBu_r', 
                      vmin=-vmax, vmax=vmax)

axes[1,2].set_title('Top 20 Most Different Reactions')
axes[1,2].set_ylabel('Method')
axes[1,2].set_xticks(np.arange(heatmap_data.shape[1]))
axes[1,2].set_xticklabels(top_diff_reactions['reaction'], rotation=90, fontsize=8)
axes[1,2].set_xlabel('Reaction')
axes[1,2].set_yticks([0, 1])
axes[1,2].set_yticklabels(['FBA', 'pFBA'])
plt.colorbar(im, ax=axes[1,2], label='Flux Value')

plt.tight_layout()
plt.show()
```
